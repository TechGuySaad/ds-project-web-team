# Noise Level Visualization

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
from datetime import datetime, timedelta
```

```python
# Load the dataset
file_path = 'Pakistan_Water_Pollution_Air_Quaility2020.csv' 
data = pd.read_csv(file_path, encoding='ISO-8859-1')

```

```python
# Step 1: Data Cleaning
# Clean column names
data.columns = data.columns.str.strip().str.replace('"', '').str.replace(' ', '')
# Clean 'City' column
data['City'] = data['City'].str.strip().str.replace('"', '')

```

```python
# Step 2: Filter data for specific cities
cities_of_interest = ['Peshawar', 'Quetta', 'Karachi', 'Lahore', 'Islamabad']
filtered_data = data[data['City'].isin(cities_of_interest)].reset_index(drop=True)

```

```python
# Step 3: Visualization
# Set the style for the visualization
sns.set(style="whitegrid")

# Create a bar plot without `palette` if no hue is specified
plt.figure(figsize=(10, 6))
sns.barplot(x='City', y='AirQuality', data=filtered_data)  

# Adding titles and labels
plt.title('Air Quality (Noise Level Proxy) in Major Cities of Pakistan', fontsize=16)
plt.xlabel('City', fontsize=14)
plt.ylabel('Air Quality Index (Lower = Noisier)', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()

# Show the plot
plt.show()
```

# Humidity Level Analysis

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Load the dataset
file_path = 'weather data cleaned 23.xlsx'  # Correct path to your uploaded file
data = pd.read_excel(file_path)


```

```python
# Step 2: Inspect the dataset
print("Dataset Columns:")
print(data.columns)
print("\nSample Data:")
print(data.head())
```

```python
# Step 3: Clean the dataset
# Clean column names
data.columns = data.columns.str.strip().str.replace(' ', '_').str.replace('\n', '')

# Define the city and humidity columns
city_col = 'Station'  # Assuming 'Station' refers to the cities or stations
humidity_col = 'Vapour_(mb)Pressure'  # Replace with the correct name if different

# Check if the required columns exist
if city_col not in data.columns or humidity_col not in data.columns:
    raise ValueError(f"The dataset must contain columns named '{city_col}' and '{humidity_col}'.")

# Step 4: Filter data for specific cities
cities_of_interest = ['Peshawar', 'Quetta', 'Karachi', 'Lahore' ,'Rawalpindi']
filtered_data = data[data[city_col].isin(cities_of_interest)].reset_index(drop=True)

# Step 5: Visualize the humidity levels
# Set the plot style
sns.set(style="whitegrid")

# Create a bar plot with 'hue' for color
plt.figure(figsize=(10, 6))
sns.barplot(x=city_col, y=humidity_col, data=filtered_data, hue=city_col, palette='coolwarm', legend=False)

# Add titles and labels
plt.title('Humidity Levels in Major Cities of Pakistan', fontsize=16)
plt.xlabel('City', fontsize=14)
plt.ylabel('Vapor Pressure (mb)', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()

# Show the plot
plt.show()
```

# As i was not having time data so inputing some data manually for time-series analysis for specific cities

```python
# Step 1: Data Preparation
# Let's assume from google that we have humidity readings from multiple locations taken over a day

np.random.seed(0)  # For reproducibility
locations = ['Lahore', 'Quetta', 'Karachi', 'Peshawar', 'Islamabad']
num_records = 48  # Number of time intervals (e.g., every 30 mins in a day)
time_stamps = [datetime(2024, 10, 30) + timedelta(minutes=30 * i) for i in range(num_records)]
humidity_data = {
    'timestamp': time_stamps * len(locations),
    'location': np.repeat(locations, num_records),
    'humidity_percent': np.random.uniform(40, 100, size=num_records * len(locations))
}

df = pd.DataFrame(humidity_data)
```

```python
# Step 2: Data Preprocessing
# Add separate columns for day and time if needed
df['time'] = df['timestamp'].dt.strftime('%H:%M')
df['date'] = df['timestamp'].dt.date
```

```python
# Step 3: Time-Series Analysis of Humidity Levels at a Specific Location
# Filtering data for Lahore
area_a_df = df[df['location'] == 'Lahore']

plt.figure(figsize=(14, 6))
plt.plot(area_a_df['timestamp'], area_a_df['humidity_percent'], marker='o', linestyle='-', color='b')
plt.title("Humidity Level Trend in Lahore")
plt.xlabel("Time")
plt.ylabel("Humidity (%)")
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```
